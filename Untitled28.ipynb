```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from functools import reduce
from itertools import cycle, islice
pd.options.display.float_format='{:.4f}'.format
plt.rcParams['figure.figsize'] = [11.5,8]
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_colwidth', 1)
```

```python
df = pd.read_csv('/content/SGDP.csv')
display(df.head())
```

```python
display(df.describe())
```

```python
df.info()
```

```python
missing_values = df.isnull().sum()
missing_values_percent = (missing_values / len(df)) * 100
display(missing_values_percent)
```

```python
df = df.dropna(axis=1, how='all')
display(df.head())
```

```python
df = df[df['Duration'] != '2016-17']
display(df.head())
```

```python
# Identify Union Territory columns
ut_columns = ['Andaman & Nicobar Islands', 'Chandigarh', 'Puducherry']

# Drop Union Territory columns
df = df.drop(columns=ut_columns)
display(df.head())
```

```python
# Select the rows corresponding to the growth percentages for the desired durations
selected_durations_df = df[df['Duration'].isin(['2013-14', '2014-15', '2015-16'])]

# Filter for rows with '% Growth over previous year'
growth_df = selected_durations_df[selected_durations_df['Items  Description'] == '(% Growth over previous year)']

# Exclude non-numeric columns before calculating the mean
numeric_growth_df = growth_df.drop(columns=['Items  Description', 'Duration'])

# Calculate the mean of the selected rows for each state
mean_growth = numeric_growth_df.mean()

# Sort the results by mean growth in ascending order
mean_growth_sorted = mean_growth.sort_values()

# Display the sorted mean growth
display(mean_growth_sorted)
```

```python
plt.rcParams['figure.figsize'] = [11.5,8]
mean_growth_sorted.plot(kind='barh',stacked=True, colormap = 'Blues_r')
plt.title("Avg.% Growth of States for Duration 2013-14, 2014-15 and 2015-16", fontweight = 'bold')
plt.xlabel("Avg. % Growth", fontweight = 'bold')
plt.ylabel("States", fontweight = 'bold')
```

```python
# Select the row corresponding to GSDP for 2015-16
dfx5_total_gdp = df[(df['Items  Description'] == 'GSDP - CURRENT PRICES (` in Crore)') & (df['Duration'] == '2015-16')]

# Drop the 'Items  Description', 'Duration', and 'All_India GDP' columns
dfx6_total_gdp = dfx5_total_gdp.drop(columns=['Items  Description', 'Duration', 'All_India GDP'])

# Transpose the DataFrame for plotting
dfx6_total_gdp = dfx6_total_gdp.T

# Rename the column to the year
dfx6_total_gdp.columns = ['2015-16']

# Plot for GSDP of all states
dfx6_total_gdp.sort_values(by='2015-16').plot(kind='bar',stacked=True, colormap = 'Blues_r')
plt.title("Total GDP of States for duration 2015-16" , fontweight = 'bold')
plt.ylabel("Total GDP (in cr)",fontweight = 'bold')
plt.xlabel("States",fontweight = 'bold')
plt.show()
```

```python
# Dropping the States whose GSDP in NaN for year 2015-16
dfx7_total_gdp = dfx6_total_gdp.dropna().sort_values(by='2015-16')

#Plot for GSDP of all states excluding States with NaN
dfx7_total_gdp.plot(kind='bar',stacked=True, colormap = 'Blues_r')
plt.title("Total GDP of States for duration 2015-16" , fontweight = 'bold')
plt.ylabel("Total GDP (in cr)",fontweight = 'bold')
plt.xlabel("States",fontweight = 'bold')
plt.show()
```

```python
# Define the regions and assign states to them
regional_groupings = {
    'North': ['Jammu & Kashmir', 'Himachal Pradesh', 'Punjab', 'Uttarakhand', 'Haryana', 'Delhi', 'Rajasthan', 'Uttar Pradesh'],
    'South': ['Andhra Pradesh', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Telangana'],
    'East': ['Bihar', 'Jharkhand', 'Odisha', 'West Bengal1'],
    'West': ['Goa', 'Gujarat', 'Maharashtra'],
    'Central': ['Madhya Pradesh', 'Chhattisgarh'],
    'North East': ['Arunachal Pradesh', 'Assam', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Tripura', 'Sikkim']
}

# Create a dictionary to map states to regions
state_to_region = {state: region for region, states in regional_groupings.items() for state in states}

# Add a 'Region' column to the DataFrame
df['Region'] = df['Items  Description'].map(state_to_region)

# Display the DataFrame with the new 'Region' column
display(df.head())
```

```python
# Filter for rows with '% Growth over previous year'
growth_df = df[df['Items  Description'] == '(% Growth over previous year)']

# Exclude non-numeric columns and 'All_India GDP' before calculating correlations with states
state_growth_df = growth_df.drop(columns=['Items  Description', 'Duration', 'All_India GDP'])

# Select the 'All_India GDP' growth column
all_india_growth = growth_df['All_India GDP']

# Calculate the correlation between each state's growth and All_India GDP growth
correlations = state_growth_df.corrwith(all_india_growth)

# Display the correlations
display(correlations.sort_values(ascending=False))
```

```python
import seaborn as sns

# Reshape the correlations for the heatmap
correlations_heatmap = correlations.to_frame(name='Correlation with All_India GDP Growth')

# Generate the heatmap
plt.figure(figsize=(8, 10))
sns.heatmap(correlations_heatmap, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation between State GSDP Growth and All_India GDP Growth', fontweight='bold')
plt.ylabel('States', fontweight='bold')
plt.show()
```
